# Back tracking:

###  5 steps:
1- start
2- choose
3- move forward
4- end point
5- go back

for each element we have a choice. we get to choose to add it or not add it. 

# Subsets
Given an array nums of unique integers, return all possible subsets of nums.

The solution set must not contain duplicate subsets. You may return the solution in any order.

Example 1:

Input: nums = [1,2,3]

Output: [[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]
Example 2:

Input: nums = [7]

Output: [[],[7]]
Constraints:

1 <= nums.length <= 10
-10 <= nums[i] <= 10


````python
def subsets(nums):
    subset = []
    res = []
    
    def dfs(i):
        if i >= len(nums):
            res.append(subset[:]) 
            print('res: ', res) # Copy the current subset
            return
        
        # include the num
        subset.append(nums[i])  # Add to subset, not res
        print('subset including the num', subset, 'i:', i)
        dfs(i+1)

        # not include the num
        subset.pop()  # Remove from subset, not res
        print('subset not including the num', subset,'i:',i)

        dfs(i+1)

    dfs(0)
    return res
    

nums = [1,2,3]
print(subsets(nums))
````

````python
# Let's trace through the algorithm step by step with i values
def subsets_with_trace(nums):
    subset = []
    res = []
    call_count = 0
    
    def dfs(i, depth=""):
        nonlocal call_count
        call_count += 1
        print(f"{depth}Call {call_count}: dfs(i={i})")
        print(f"{depth}Current subset: {subset}")
        print(f"{depth}i={i}, nums length={len(nums)}")
        
        if i >= len(nums):
            print(f"{depth}★ BASE CASE: i={i} >= {len(nums)}, adding {subset[:]} to result")
            res.append(subset[:])  
            print(f"{depth}Result so far: {res}")
            print(f"{depth}Returning from dfs(i={i})")
            return
        
        print(f"{depth}Processing nums[{i}] = {nums[i]}")
        
        # INCLUDE the current element
        print(f"{depth}► CHOICE 1: Include nums[{i}] = {nums[i]}")
        subset.append(nums[i])  
        print(f"{depth}  Subset after adding: {subset}")
        print(f"{depth}  Calling dfs(i={i+1})")
        dfs(i+1, depth + "    ")  # Recursive call with i+1
        
        # BACKTRACK - remove the element
        print(f"{depth}◄ BACKTRACKING from dfs(i={i+1})")
        print(f"{depth}  Removing nums[{i}] = {nums[i]} from subset")
        subset.pop()  
        print(f"{depth}  Subset after removing: {subset}")
        
        # DON'T INCLUDE the current element
        print(f"{depth}► CHOICE 2: Don't include nums[{i}] = {nums[i]}")
        print(f"{depth}  Calling dfs(i={i+1}) without adding element")
        dfs(i+1, depth + "    ")  # Recursive call with i+1
        
        print(f"{depth}◄ Finished both choices for nums[{i}] = {nums[i]}")

    print("=== STARTING BACKTRACKING ===")
    print(f"Input array: {nums}")
    dfs(0)
    return res

# Run the traced version
nums = [1, 2]  # Using smaller example for clarity
result = subsets_with_trace(nums)
print("\n" + "="*50)
print("FINAL RESULT")
print("="*50)
print(f"All subsets: {result}")
print(f"Total subsets found: {len(result)}")
print(f"Expected: 2^{len(nums)} = {2**len(nums)} subsets")
````

## Understanding the `i` parameter

The parameter `i` represents the **current index** we're considering in the `nums` array.

| Call | i | Action | Subset State | Meaning |
|------|---|--------|-------------|---------|
| 1 | 0 | Start | `[]` | Consider nums[0] = 1 |
| 2 | 1 | Include 1 | `[1]` | Consider nums[1] = 2 |
| 3 | 2 | Include 2 | `[1,2]` | i >= len(nums), BASE CASE |
| 4 | 2 | Backtrack, exclude 2 | `[1]` | i >= len(nums), BASE CASE |
| 5 | 1 | Backtrack, exclude 1 | `[]` | Consider nums[1] = 2 |
| 6 | 2 | Include 2 | `[2]` | i >= len(nums), BASE CASE |
| 7 | 2 | Exclude 2 | `[]` | i >= len(nums), BASE CASE |

**Key Points:**
- `i` starts at 0 (first element)
- Each recursive call increments `i` by 1
- When `i >= len(nums)`, we've considered all elements → BASE CASE
- The algorithm explores all 2^n possible combinations

## How Backtracking Works - Visual Tree

For `nums = [1, 2]`, the algorithm explores this decision tree:

```
                    dfs(0) - subset=[]
                   /                \
              Include 1                Don't include 1
             subset=[1]                   subset=[]
                 |                           |
            dfs(1) - subset=[1]        dfs(1) - subset=[]
           /              \           /              \
    Include 2        Don't include 2    Include 2    Don't include 2
   subset=[1,2]        subset=[1]      subset=[2]      subset=[]
        |                  |              |              |
   dfs(2) BASE        dfs(2) BASE    dfs(2) BASE    dfs(2) BASE
   Add [1,2]          Add [1]        Add [2]        Add []
```

**Key Steps:**
1. **Choose**: At each node, decide to include or exclude current element
2. **Move Forward**: Make recursive call with next index
3. **End Point**: When `i >= len(nums)`, add current subset to result
4. **Backtrack**: Remove element and try the other choice

**Order of subsets added:** `[]`, `[2]`, `[1]`, `[1,2]`
