# Reverse LinkedList (using recursive)

````python
def reverseList(head):
    if not head:
        return None
    itr = head

    if head.next:
        itr = reverseList(head.next)
        head.next.next = head
    head.next = None
    return itr

````

## Understanding Recursive Linked List Reversal

This is one of the trickiest recursions to understand! Let's break it down step by step.

### **The Algorithm:**
```python
def reverseList(head):
    if not head:           # Edge case: empty list
        return None
    itr = head            # itr will track the new head
    
    if head.next:         # If there's more than one node
        itr = reverseList(head.next)  # Recursively reverse the rest
        head.next.next = head         # Reverse the current connection
    head.next = None      # Break the old connection
    return itr           # Return the new head
```

### **Key Variables:**
- `head`: Current node we're processing
- `itr`: The new head of the reversed list (gets passed back up)
- `head.next`: The next node in the original list

````python
# Let's trace through the recursion step by step
def reverseList_trace(head, depth=0):
    indent = "  " * depth
    node_val = head.val if head else "None"
    print(f"{indent}📞 Call {depth + 1}: reverseList({node_val})")
    
    if not head:
        print(f"{indent}❌ Base case: head is None, returning None")
        return None
    
    print(f"{indent}📍 Current node: {head.val}")
    itr = head
    print(f"{indent}🎯 itr initially points to: {itr.val}")
    
    if head.next:
        next_val = head.next.val if head.next else "None"
        print(f"{indent}🔄 head.next exists ({next_val}), going deeper...")
        
        # The recursive magic happens here
        itr = reverseList_trace(head.next, depth + 1)
        
        print(f"{indent}🔙 Back from recursion! itr now points to: {itr.val}")
        print(f"{indent}📝 Current state:")
        print(f"{indent}   - head: {head.val}")
        print(f"{indent}   - head.next: {head.next.val}")
        print(f"{indent}   - itr (new head): {itr.val}")
        
        # The key reversal step
        print(f"{indent}🔀 Reversing: {head.next.val}.next = {head.val}")
        head.next.next = head
        
        print(f"{indent}📎 Now {head.next.val} points back to {head.val}")
    else:
        print(f"{indent}🏁 Base case: {head.val} is the last node (new head)")
    
    print(f"{indent}✂️  Setting {head.val}.next = None (breaking old connection)")
    head.next = None
    
    print(f"{indent}📤 Returning new head: {itr.val}")
    return itr

# Helper to create a linked list for testing
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def print_linked_list(head):
    result = []
    current = head
    while current:
        result.append(str(current.val))
        current = current.next
    return " → ".join(result) if result else "None"

# Test the recursive reversal
print("=== RECURSIVE LINKED LIST REVERSAL TRACE ===")
test_list = create_linked_list([1, 2, 3])
print(f"Original list: {print_linked_list(test_list)}")
print("\n--- Starting recursion ---")
reversed_head = reverseList_trace(test_list)
print(f"\n🎉 Final reversed list: {print_linked_list(reversed_head)}")
````

## Visual Step-by-Step for [1, 2, 3]

Let's trace through `reverseList([1,2,3])`:

### **Call Stack (Going Down):**
```
Call 1: reverseList(1) → head=1, head.next=2
  ├─ Call 2: reverseList(2) → head=2, head.next=3  
      ├─ Call 3: reverseList(3) → head=3, head.next=None
      │   └─ Base case! itr=3, return 3
      └─ Back to Call 2: itr=3 (from Call 3)
  └─ Back to Call 1: itr=3 (from Call 2)
```

### **Unwinding (Coming Back Up):**

**Call 3 (Base case):**
- `head = 3`, `head.next = None`
- `itr = 3` (this becomes our new head!)
- Return `3`

**Call 2 (Unwinding):**
- `head = 2`, `head.next = 3`
- `itr = 3` (received from Call 3)
- **Key step:** `head.next.next = head` → `3.next = 2`
- **Break old:** `head.next = None` → `2.next = None`
- Return `3`

**Call 1 (Final):**
- `head = 1`, `head.next = 2` 
- `itr = 3` (received from Call 2)
- **Key step:** `head.next.next = head` → `2.next = 1`
- **Break old:** `head.next = None` → `1.next = None`
- Return `3`

### **The Magic Line:**
`head.next.next = head` is where the reversal happens!

- In Call 2: `2.next.next = 2` → `3.next = 2`
- In Call 1: `1.next.next = 1` → `2.next = 1`

This creates: `3 → 2 → 1 → None`

## 🔗 How Connections Are Maintained

The key insight is that **each recursive call maintains its own local connections**, and the magic happens when they all combine!

### **The Connection Strategy:**

1. **`itr` carries the new head up the call stack**
2. **Each level reverses only its immediate connection**
3. **`head.next = None` prevents cycles and breaks old connections**

### **Connection Flow for [1,2,3]:**

```
Before recursion: 1 → 2 → 3 → None

During recursion:
┌─────────────────────────────────────┐
│ Call 1: head=1                      │
│ ├─ itr = reverseList(2→3)          │
│ │   ┌─────────────────────────────┐ │
│ │   │ Call 2: head=2              │ │
│ │   │ ├─ itr = reverseList(3)     │ │  
│ │   │ │   ┌─────────────────────┐ │ │
│ │   │ │   │ Call 3: head=3      │ │ │
│ │   │ │   │ itr = 3             │ │ │
│ │   │ │   │ return 3            │ │ │
│ │   │ │   └─────────────────────┘ │ │
│ │   │ │                           │ │
│ │   │ ├─ itr = 3                  │ │
│ │   │ ├─ 3.next = 2 (REVERSE!)    │ │
│ │   │ ├─ 2.next = None            │ │
│ │   │ └─ return 3                 │ │
│ │   └─────────────────────────────┘ │
│ │                                   │
│ ├─ itr = 3                          │
│ ├─ 2.next = 1 (REVERSE!)            │
│ ├─ 1.next = None                    │
│ └─ return 3                         │
└─────────────────────────────────────┘

Final result: 3 → 2 → 1 → None
```

````python
# Let's trace the connection maintenance in detail
def reverseList_connection_trace(head, depth=0):
    indent = "  " * depth
    node_val = head.val if head else "None"
    print(f"{indent}🔄 CALL {depth + 1}: Processing node {node_val}")
    
    if not head:
        print(f"{indent}❌ Empty node - returning None")
        return None
    
    print(f"{indent}📍 Current connections:")
    next_val = head.next.val if head.next else "None"
    print(f"{indent}   {head.val} → {next_val}")
    
    itr = head
    print(f"{indent}🎯 itr initially = {itr.val}")
    
    if head.next:
        print(f"{indent}🏃 Going deeper to process {head.next.val}...")
        
        # THE MAGIC: Recursion processes the rest
        itr = reverseList_connection_trace(head.next, depth + 1)
        
        print(f"{indent}🔙 BACK! Recursion returned itr = {itr.val}")
        print(f"{indent}🔗 Current state before reversing:")
        print(f"{indent}   head = {head.val}")
        print(f"{indent}   head.next = {head.next.val}")
        print(f"{indent}   itr (new head) = {itr.val}")
        
        # THE KEY REVERSAL STEP
        print(f"{indent}⚡ REVERSING: {head.next.val}.next = {head.val}")
        head.next.next = head
        
        # Show the connection after reversal
        print(f"{indent}✅ After reversal:")
        print(f"{indent}   {head.next.val} → {head.val}")
        
    else:
        print(f"{indent}🏁 BASE CASE: {head.val} is the last node (becomes new head)")
    
    # Breaking the old connection
    print(f"{indent}✂️  BREAKING old connection: {head.val}.next = None")
    head.next = None
    
    print(f"{indent}📤 Returning new head: {itr.val}")
    print(f"{indent}🔍 Current chain from {itr.val}:")
    
    # Show current state of the chain
    current = itr
    chain = []
    visited = set()
    while current and current.val not in visited:
        chain.append(str(current.val))
        visited.add(current.val)
        current = current.next
        if len(chain) > 5:  # Prevent infinite loops
            break
    
    print(f"{indent}   {' → '.join(chain)} → ...")
    return itr

# Test with a simple example
print("=== CONNECTION MAINTENANCE TRACE ===")
test_list = create_linked_list([1, 2, 3])
print(f"Original: {print_linked_list(test_list)}")
print("\n--- Tracing Connections ---")
result = reverseList_connection_trace(test_list)
print(f"\n🎉 FINAL: {print_linked_list(result)}")
````

## 🧠 The Mental Model for Connection Keeping

Think of it like a **chain of trust**:

### **Step 1: Trust the Recursion**
```
reverseList(1→2→3)
├─ "I'll handle node 1, but first..."
└─ "Hey reverseList(2→3), you handle the rest!"
```

### **Step 2: Each Call's Responsibility**
```python
# Each call only worries about:
itr = reverseList(head.next)  # "Handle everything after me"
head.next.next = head         # "Reverse just MY connection"
head.next = None             # "Break MY old connection"
return itr                   # "Pass up the new head"
```

### **Step 3: The Connection Magic**

**Key insight:** `itr` is like a **baton in a relay race**
- It carries the "new head" up through all recursive calls
- Each level adds their reversed connection to the growing chain
- The final `itr` points to the fully reversed list

### **Why `head.next = None` is Critical:**

Without this line, you'd get **cycles**:
```
❌ Without head.next = None:
1 ⟷ 2 → 3   (1 and 2 point to each other!)

✅ With head.next = None:
1 ← 2 ← 3   (clean reversal)
```

### **The Beautiful Pattern:**
1. **Recurse first** (handle the "rest")
2. **Reverse locally** (just your immediate connection)  
3. **Clean up** (break old connections)
4. **Pass up the head** (maintain the chain)

This way, each call maintains **exactly the connections it's responsible for**, and they all combine into the final reversed list!

#  Swap Nodes in Pairs

Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)

 

Example 1:

Input: head = [1,2,3,4]

Output: [2,1,4,3]

Output: []

Example 3:

Example 4:

Input: head = [1,2,3]

Output: [2,1,3]

 

Constraints:

The number of nodes in the list is in the range [0, 100].
0 <= Node.val <= 100

````python
def swapPairs(head):
    # Handle edge cases
    if not head or not head.next:
        return head
    
    # Create a dummy node to simplify logic
    dummy = Node(0)
    dummy.next = head
    prev = dummy
    
    # Process pairs
    while prev.next and prev.next.next:
        # Identify the two nodes to swap
        first = prev.next
        second = prev.next.next
        
        # Perform the swap
        prev.next = second
        first.next = second.next
        second.next = first
        
        # Move prev to the end of swapped pair
        prev = first
    
    return dummy.next

````

````python
# Alternative simpler recursive approach
def swapPairs_recursive(head):
    if not head or not head.next:
        return head
    
    # Save the second node
    second = head.next
    
    # Recursively swap the rest
    head.next = swapPairs_recursive(second.next)
    
    # Swap first two nodes
    second.next = head
    
    return second
````

## Recursive Approach Explained

The recursive approach uses the "divide and conquer" strategy:

### **Core Idea:**
1. **Base case**: If less than 2 nodes, return as-is
2. **Recursive case**: 
   - Swap the first two nodes
   - Recursively handle the rest of the list
   - Connect the swapped pair to the processed rest

### **Visual Example:**
For list `[1, 2, 3, 4]`:

```
Call 1: swapPairs([1,2,3,4])
├── Base case? No (has 2+ nodes)
├── second = node(2)
├── Recursively process [3,4] → returns [4,3]
├── Connect: 1 → [4,3] = [1,4,3]
└── Swap: 2 → 1 → [4,3] = [2,1,4,3]

Call 2: swapPairs([3,4]) 
├── Base case? No (has 2+ nodes)
├── second = node(4)
├── Recursively process [] → returns []
├── Connect: 3 → [] = [3]
└── Swap: 4 → 3 = [4,3]

Call 3: swapPairs([])
└── Base case? Yes → return []
```
