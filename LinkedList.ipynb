#  Swap Nodes in Pairs

Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)

 

Example 1:

Input: head = [1,2,3,4]

Output: [2,1,4,3]

Explanation:



Example 2:

Input: head = []

Output: []

Example 3:

Input: head = [1]

Output: [1]

Example 4:

Input: head = [1,2,3]

Output: [2,1,3]

 

Constraints:

The number of nodes in the list is in the range [0, 100].
0 <= Node.val <= 100

````python
def swapPairs(head):
    # Handle edge cases
    if not head or not head.next:
        return head
    
    # Create a dummy node to simplify logic
    dummy = Node(0)
    dummy.next = head
    prev = dummy
    
    # Process pairs
    while prev.next and prev.next.next:
        # Identify the two nodes to swap
        first = prev.next
        second = prev.next.next
        
        # Perform the swap
        prev.next = second
        first.next = second.next
        second.next = first
        
        # Move prev to the end of swapped pair
        prev = first
    
    return dummy.next

````

````python
# Alternative simpler recursive approach
def swapPairs_recursive(head):
    if not head or not head.next:
        return head
    
    # Save the second node
    second = head.next
    
    # Recursively swap the rest
    head.next = swapPairs_recursive(second.next)
    
    # Swap first two nodes
    second.next = head
    
    return second
````

## Recursive Approach Explained

The recursive approach uses the "divide and conquer" strategy:

### **Core Idea:**
1. **Base case**: If less than 2 nodes, return as-is
2. **Recursive case**: 
   - Swap the first two nodes
   - Recursively handle the rest of the list
   - Connect the swapped pair to the processed rest

### **Visual Example:**
For list `[1, 2, 3, 4]`:

```
Call 1: swapPairs([1,2,3,4])
├── Base case? No (has 2+ nodes)
├── second = node(2)
├── Recursively process [3,4] → returns [4,3]
├── Connect: 1 → [4,3] = [1,4,3]
└── Swap: 2 → 1 → [4,3] = [2,1,4,3]

Call 2: swapPairs([3,4]) 
├── Base case? No (has 2+ nodes)
├── second = node(4)
├── Recursively process [] → returns []
├── Connect: 3 → [] = [3]
└── Swap: 4 → 3 = [4,3]

Call 3: swapPairs([])
└── Base case? Yes → return []
```

## Step-by-Step Breakdown

Let's analyze each line of the recursive function:

```python
def swapPairs_recursive(head):
    # Step 1: Base case - stop recursion
    if not head or not head.next:
        return head
    
    # Step 2: Save reference to second node
    second = head.next
    
    # Step 3: Recursively process the rest
    head.next = swapPairs_recursive(second.next)
    
    # Step 4: Perform the swap
    second.next = head
    
    # Step 5: Return new head of this segment
    return second
```

### **Why this works:**

1. **Trust the recursion**: Assume `swapPairs_recursive(second.next)` correctly swaps all remaining pairs
2. **Focus on current pair**: Only worry about swapping the first two nodes
3. **Connect properly**: Link the swapped pair to the already-processed rest
4. **Return new head**: The second node becomes the new head of this segment

### **Key insight:**
The recursion handles the "rest of the list" for us, so we only need to:
- Swap the first two nodes  
- Connect them to whatever the recursion returns

This is much simpler than the iterative approach because we don't need to manually track all the connections!

2. Use consistent swapping logic for each pair

For every pair, follow the same 3-step process:

Save references to both nodes
Rewire the connections
Move to the next pair

3. Handle connections between

# The systematic approach in code:
Key aspects of "systematic processing":

1. Consistent Pattern

Each pair follows the exact same swapping steps:

Identify first and second nodes
Save reference to what comes after
Rewire connections in the same order
Move to next pair

2. Proper State Management

Use a prev pointer to track where we are
Always know what comes before and after each pair
Maintain connections between processed and unprocessed parts

3. Edge Case Handling

Check if pair exists before processing
Handle odd number of nodes gracefully
Use dummy node to simplify boundary conditions

4. No Manual Tracking

Instead of trying to remember "which nodes did I swap?", the algorithm automatically processes each pair in sequence using the while loop condition.
